```python
class Person:
    def __init__(self, name, age, location):
        self.name = name
        self.age = age
        self.location = location

    def greet(self):
        print(f"Hello, my name is {self.name}.")

class Student(Person):
    def __init__(self, name, age, location, student_id):
        super().__init__(name, age, location)
        self.student_id = student_id

    def greet(self):
        print(f"Hi, I'm {self.name} and my student ID is {self.student_id}.")

def display_info(person):
    print(f"Name: {person.name}")
    print(f"Age: {person.age}")
    print(f"Location: {person.location}")
    if isinstance(person, Student):
        print(f"Student ID: {person.student_id}")
    print()

# Instantiate objects
person1 = Person("John Doe", 25, "City A")
person2 = Person("Jane Smith", 30, "City B")
person3 = Person("Bob Johnson", 22, "City C")

student1 = Student("Alice Williams", 21, "City D", "S12345")
student2 = Student("Charlie Brown", 23, "City E", "S67890")

# Test greet method
person1.greet()
student1.greet()

# Display information
display_info(person2)
display_info(student2)
```
