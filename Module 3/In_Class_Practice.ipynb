**In-Class Practice Assignment: Static Methods**

**Objective:**
- Understand the concept of static methods in Python and their use in class design.

**Instructions:**
1. Create a class named `Calculator`.
2. Implement two instance methods: `add` and `subtract`. These methods should take two parameters and perform addition and subtraction, respectively.
3. Implement a static method named `multiply` that takes two parameters and performs multiplication without the need for creating an instance of the class.
4. Create instances of the `Calculator` class and use the `add` and `subtract` instance methods to perform calculations.
5. Use the `multiply` static method without creating an instance of the class to perform multiplication.
6. Display the results of the calculations.



This assignment focuses on static methods in Python, demonstrating their usage in a class and how they can perform operations without relying on instance-specific attributes. Students can practice creating static methods, calling them without creating instances, and understanding their role in class design.

````python
class Calculator:
    def add(self, num1, num2):
        return num1 + num2

    def subtract(self, num1, num2):
        return num1 - num2

    @staticmethod
    def multiply(num1, num2):
        return num1 * num2

# Creating instances
calculator1 = Calculator()
calculator2 = Calculator()

# Using instance methods for addition and subtraction
result_add = calculator1.add(10, 5)
result_subtract = calculator2.subtract(15, 7)

# Using static method for multiplication
result_multiply = Calculator.multiply(3, 4)

# Displaying results
print(f"Addition Result: {result_add}")
print(f"Subtraction Result: {result_subtract}")
print(f"Multiplication Result: {result_multiply}")
```
````
