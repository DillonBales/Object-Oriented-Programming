**Problem 1: Class Attributes and Methods**

**Objective:**
- Understand the concept of class attributes and methods in Python.

**Instructions:**
1. Create a class called `Book` with a class attribute `genre` set to "Fiction."
2. Define an `__init__` method to initialize instance attributes `title` and `author`.
3. Implement a class method named `change_genre` that allows changing the genre for all books.
4. Create instances of the `Book` class with different titles and authors.
5. Access the class attribute and instance attributes for each book.
6. Use the `change_genre` class method to update the genre for all books.
7. Display the updated genre and instance attributes for each book.





**Problem 2: Instance Methods vs Class Methods**

**Objective:**
- Differentiate between instance methods and class methods in Python.

**Instructions:**
1. Create a class named `Rectangle` with instance attributes `length` and `width`.
2. Implement an instance method `calculate_area` that calculates and returns the area of the rectangle.
3. Implement a class method named `create_square` that creates a square by setting both length and width to the same value.
4. Create instances of the `Rectangle` class with different lengths and widths.
5. Use the `calculate_area` instance method to display the area of each rectangle.
6. Use the `create_square` class method to create a square and display its area.



These assignment provide hands-on experience with class attributes, instance methods, and class methods in Python, reinforcing the understanding of object-oriented programming concepts.



### **Problem 3: Static Methods**

#### **Objective:**
Understand the concept of static methods in Python.

#### **Instructions:**

1. Create a class named `TemperatureConverter` with a static method named `celsius_to_fahrenheit` that converts a temperature from Celsius to Fahrenheit. The formula for conversion is Fahrenheit = (Celsius * 9/5) + 32.

2. Implement another static/class method named `fahrenheit_to_celsius` that converts a temperature from Fahrenheit to Celsius. The formula for conversion is Celsius = (Fahrenheit - 32) * 5/9.

3. Create instances of the `TemperatureConverter` class and test both static methods for temperature conversion.



### **Problem 4: Class Inheritance**

#### **Objective:**
Explore class inheritance in Python.

#### **Instructions:**

1. Create a base class named `Animal` with instance attributes `name` and `sound`. Implement an instance method named `make_sound` that prints the sound of the animal.

2. Create two derived classes, `Dog` and `Cat`, that inherit from the `Animal` class. Add a new instance attribute to each derived class - `breed` for `Dog` and `fur_color` for `Cat`.

3. Override the `make_sound` method in both `Dog` and `Cat` classes to include the sound and an additional characteristic (e.g., breed for dogs, fur color for cats).

4. Create instances of the `Dog` and `Cat` classes, set values for attributes, and call the `make_sound` method for each.
